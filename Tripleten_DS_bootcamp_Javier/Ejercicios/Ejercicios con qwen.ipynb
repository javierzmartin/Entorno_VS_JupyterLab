```python
frases = [
    ["Hola", "mundo"],       # 2 palabras
    ["Python", "es", "genial"],  # 3 palabras
    ["Aprendiendo", "Python"],   # 2 palabras
    ["Python", "es", "genial"]   # 3 palabras (empate)
]

# C칩digo actual (con `if num_palabras > max_palabras`)
max_palabras = 0
frase_max = None

for frase in frases:
    num_palabras = len(frase)
    if num_palabras > max_palabras:
        max_palabras = num_palabras
        frase_max = frase

print(frase_max)  # Salida: ["Python", "es", "genial"] (la **segunda** frase, no la 칰ltima)
```
