```python
# Pido el input
tareas = input('Dame una lista de tareas (puedes separarlas con , ; √≥ /')
# Deteci√≥n del separador, cambia el sepador por uno gen√©rico, en este caso una coma
tareas_modificadas = tareas.replace(';',',').replace('/',',')
# Convierto a una lista
tareas_modificadas = tareas_modificadas.split(',')
# Quita espacios
for i in range(len(tareas_modificadas)):
    tareas_modificadas[i]=tareas_modificadas[i].strip()


for i,tarea in enumerate(tareas_modificadas,start=1):
    print(f"{i}) {tarea}")
```

## üéØ RETO 2 ‚Äî Limpieza avanzada de la lista de tareas

üß© Resumen del reto
Tu programa debe:

Todo con un estilo paso a paso, sin atajos raros, sin comprensiones de lista si no quieres.




```python
lista = input('Dame una lista separada por comas')
# Convierto los caracteres en minusculas
lista_modificada = lista.lower()
# Convierto en lista dividiendo por la coma
lista_modificada = lista_modificada.split(',')
# Quito espacios de los lados de cada elemento
lista_modificada_sin_espacios = []
for i in range(len(lista_modificada)):
    elemento = lista_modificada[i].strip()
    lista_modificada_sin_espacios.append(elemento)
# Si un elemento esta vacio se saca de la lista
lista_sin_elementos_vacios = []
for i in range(len(lista_modificada_sin_espacios)):
    elemento = lista_modificada_sin_espacios[i]
    if elemento != '':
        lista_sin_elementos_vacios.append(elemento)
# Si hay un elemento repetido lo quita de la lista
lista_formateada =[]
for elemento in lista_sin_elementos_vacios:
    if elemento not in lista_formateada:
        lista_formateada.append(elemento)
# Enumera la lista, con el indice desplazado a 1 para que sea legible como un podio
for i,elemento in enumerate(lista_formateada, start=1):
    print(f'{i}) {elemento}')
```
