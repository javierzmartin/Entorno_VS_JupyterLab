```python
import pandas as pd 

df = pd.read_csv('C:/Users/becem/Desktop/Entorno_VS_Jupiterlab/Datasets/valentines_day_spending_dataset.csv')

# la función describe con el argumento include='' da valores
# sobre el tipo de dato que se indique, object en este caso.

# print(df.describe(include = 'object'))

# El kernel me indica que será sustitudo por str en un fúturo lo aplico así
print(df.describe(include = 'str'))

# Count es el numero de datos
# Unique la cantidad de datos diferentes para ese valor
# Top indica cual es el valor que se repite más
# Frecuency la cantidad de veces que del valor más repetido

# En esta petición pido numeros float
print(df.describe(include = 'float'))

# Mean es la media de valores del dato
# Std es la desviación tipica (la diferencia media más común entre datos)
# Min indica el valor minimo dentro de la clumna
# Max indica el valor máximo
# los percentiles indican que el x% de los valores se encuentran por debajo del valor indicado.
```
