# Procesamiento de valores duplicados

Duplicados explicitos (coincidencias en todas las columnas)

```python
import pandas as pd 
df = pd.read_csv('C:/Users/becem/Desktop/Entorno_VS_Jupiterlab/Datasets/valentines_day_spending_dataset.csv')
# Saltaba error de definición para df_esp en esta celda incluyo esta llamada al cambio de nombres
df_esp = df.rename(columns=columns_esp)

print(df_esp.duplicated().sum())

# drop_duplicates elimina las filas duplicadas

df_esp.drop_duplicates(inplace=True)

# reset_index reasigna indices
df_esp.reset_index(drop=True)
```

# Gráfica de dispersión para age y spending_usd

```python
# llamo a pyplot para genrerar gráficas
from matplotlib import pyplot as plt

df.plot(x = 'age',
        y = 'spending_usd',
        kind = 'scatter',
        title = 'Adult heights', 
        alpha = 0.36, 
        figsize = [8,6], 
        xlabel = 'Age / years', 
        ylabel = 'Spending / $')

print(df['age'].corr(df['spending_usd']))

plt.show()
```

- Hay una muy ligera desviación negativa para la correlación entre la edad y el gasto.
- Refleja una muy ligera tendencia a un gasto mayor a edades más tempranas.
