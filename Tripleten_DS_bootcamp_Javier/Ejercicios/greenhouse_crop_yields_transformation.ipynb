Este notebook incluye un proceso de limpieza y transformación que da lugar a una versión derivada del dataset.

```python
import pandas as pd

df = pd.read_csv('C:/Users/becem/Desktop/Entorno_VS_Jupiterlab/Datasets/greenhouse_crop_yields.csv')
df.info()
```

Observo que hay datos nulos en algunas columnas, no parece que por la cantidad vaya a ser un problema sacar conclusiones acertadas, de todas formas hay que quitarlos.

```python
# Creo una lista de columnas con datos nulos
nan = ['avg_temperature_C',
       'min_temperature_C',
       'max_temperature_C',
       'humidity_percent',
       'co2_ppm',
       'light_intensity_lux',
       'fertilizer_N_kg_ha',
       'fertilizer_P_kg_ha',
       'fertilizer_K_kg_ha',
       'pest_severity',
      'soil_pH']
for c in nan:
    df = df.dropna(subset = c)
df.info()
```

He perdido aproximadamente 2000 muestras no esta del todo mal

```python
# Voy a eliminar duplicados

df = df.drop_duplicates()
df.info()
```

Voy a ver la estructura y forma de los datos str

```python
df.describe(include = 'str')
```

Quiero ver si hay duplicados implicitos, no creo porque parecen bastante fijos por el tamaño, pero nunca está de más

```python
print(df['crop_type'].sort_values().unique())
print()
print(df['variety'].sort_values().unique())
```

Todo bien, solo recordar que tienen letra capital.

Voy a ver la cantidad de invernaderos que hay.

```python
print(df['greenhouse_id'].sort_values().unique())
```

Son numeros enteros, no es necesario tratarlos como float

```python
df['greenhouse_id'] = df['greenhouse_id'].astype('int')
print(df['greenhouse_id'].info())
```

El dato days_to_maturity es float pero son numeros enteros

```python
print(df['days_to_maturity'].sample(5))
```

```python
df['days_to_maturity'] = df['days_to_maturity'].astype('int')
print(df['days_to_maturity'].info())
```

Voy a tratar las fechas de siembra y recogida, las voy a convertir en tipo datetime

```python
fechas = ['planting_date','harvest_date']
# No hace falta pero lo hice con un for
for c in fechas:
    df[c] = pd.to_datetime(df[c], format = '%Y-%m-%d')
    print(df[c].info())
    print()
```

Verifico la estructura de los datos

```python
df.info()
```

Elimino la columna days_to_maturity voy a calcular más adelante eso con los datos de fechas que tengo

```python
df = df.drop('days_to_maturity', axis=1)
df.info()
```

```python
# Con las fechas convertidas a datetime se pueden hacer cálculos directos
df['days_to_maturity'] = df['harvest_date'] - df['planting_date']
df.info()
```

Para ver la estructura mejor cambio el indice de days_to_maturity

```python
df.insert(5, "days_to_maturity", df.pop('days_to_maturity'))
df.info()
```

```python
df.to_csv("greenhouse_crop_yields_transformation.csv", index=False)
```
