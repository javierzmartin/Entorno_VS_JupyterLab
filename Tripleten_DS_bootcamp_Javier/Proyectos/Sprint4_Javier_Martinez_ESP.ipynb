````python
# Creo un diccionario para aplicar en la función que lea todos los csv e indique separador punto y coma
# Asigné nombres alternativos para que no se solapen con los que voy a trabajar
csv_dic_alter = {'df_instacart_orders_alter' : '/datasets/instacart_orders.csv',
            'df_instacart_products_alter': '/datasets/products.csv',
            'df_aisles_alter' : '/datasets/aisles.csv',
            'df_departments_alter' : '/datasets/departments.csv',
            'df_order_products_alter' : '/datasets/order_products.csv'}
# Defino la función con el parametro diccionario
def read_csv_sep_semicolon(diccionario):
    # Itero sobre (clave, ruta) que devuelve diccionario.items()
    for clave, ruta in diccionario.items():
        # Indico el valor de cada data frame con (clave, ruta)
        diccionario[clave] = pd.read_csv(ruta, sep=";")

    for clave, df in diccionario.items():
        # Imprimo los resultados
        print(f'.....INFO {clave.upper()}.....')
        df.info()
        print()
    return 
read_csv_sep_semicolon(csv_dic_alter)
````
